# Geocode January 2023 FFLs
This data will be useful for calculating the driving distances

```python
df = pd.read_csv('../../data/processed/atf-ffl-list/2023-ffl-list-clean.csv')
```

```python
# Create full address column
df["original_address"] = df["premise_street"] + ", " + df["premise_city"] + ", " + df["premise_state"] + ", "  + df["premise_zip_code"]
```

```python
batch = df[["original_address"]].copy()
```

```python
batch = batch.drop_duplicates().reset_index(drop=True)
```

```python
batch['longitude'] = ""
batch['latitude'] = ""
batch['clean_address'] = ""
```

```python
batch.head()
```

```python
# Specify which API key to use
MAPS_KEY = os.getenv('MAPS_KEY')
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json?'
```

```python
for x in range(len(batch)):
    place = batch['original_address'][x]
    print(x, ":", place)
    try:
        geocode_result = requests.get(url + 'address=' + place + '&key=' + MAPS_KEY).json()
        batch['longitude'][x] = geocode_result.get('results')[0].get('geometry').get('location').get('lng')
        batch['latitude'][x] = geocode_result.get('results')[0].get('geometry').get('location').get('lat')
        batch['clean_address'][x] = geocode_result.get('results')[0].get('formatted_address')
    except:
        batch['longitude'][x] = ""
        batch['latitude'][x] = ""
        batch['clean_address'][x] = ""
```
